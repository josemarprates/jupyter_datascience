#### Agora vamos analisar as variáveis categoricas usando Teste estatistico Qi-Quadrado ####

```python
from scipy.stats import chi2_contingency

# Selecionar as variáveis categóricas (excluindo a variável 'Fraude')
variaveis_categoricas = df.select_dtypes(include=['object']).columns.tolist()

# Criar um DataFrame para armazenar os resultados
resultados_chi2 = pd.DataFrame(columns=['Variável', 'Qui-Quadrado', 'P-valor', 'Significância'])

# Calcular o teste qui-quadrado para cada variável categórica
for variavel in variaveis_categoricas:
    # Criar uma tabela de contingência
    tabela_contingencia = pd.crosstab(df[variavel], df['Fraude'])
    
    # Aplicar o teste qui-quadrado
    chi2, p, dof, expected = chi2_contingency(tabela_contingencia)
    
    # Determinar se o p-valor é significativo
    significancia = 'Sim' if p < 0.05 else 'Não'
    
    # Armazenar os resultados
    resultados_chi2 = resultados_chi2.append({
        'Variável': variavel,
        'Qui-Quadrado': chi2,
        'P-valor': p,
        'Significância': significancia
    }, ignore_index=True)

# Exibir os resultados
resultados_chi2.sort_values(by='P-valor')
```

### DECIDI EM TOMAR OUTRO RUMO EM MINHA ANÁLISE EM VERIFICAR A COMBINAÇÃO DE VARIÁVEIS ###

A seguir vou para as prxoximas etapas da análise:

Diante dos testes estatísticos que indicam que não há correlação significativa entre as variáveis categóricas e a variável alvo Fraude, e com correlações fracas ou inexistentes também nas variáveis numéricas, sugiro algumas abordagens que você pode considerar para melhorar a capacidade preditiva do modelo:

1. Criação de Novas Variáveis (Feature Engineering):
Interações entre Variáveis: Criar novas variáveis que representem interações entre as variáveis existentes. Por exemplo, a combinação de Valor_Transacao e Tentativas_Autenticacao_Falhas pode capturar comportamentos suspeitos que não são visíveis ao olhar essas variáveis individualmente.
Variáveis Temporais: Extraia características temporais a partir de Data_Transacao, como o dia da semana, hora do dia, ou se a transação ocorreu durante um feriado. Fraudes podem ser mais prováveis em horários ou dias específicos.
Diferenças entre Transações: Calcular diferenças entre transações consecutivas no mesmo cartão ou dentro do mesmo local, como a diferença de tempo ou valor entre transações. Isso pode identificar comportamentos anômalos.

2. Utilização de Modelos Avançados:
Modelos Não Lineares: Modelos como árvores de decisão, florestas aleatórias (Random Forest), ou gradiente boosting (XGBoost) podem capturar relações complexas e não lineares entre as variáveis que um modelo linear (como regressão logística) não conseguiria.
Redes Neurais: Se houver um volume significativo de dados, redes neurais podem capturar padrões complexos que são difíceis de detectar com métodos tradicionais.

3. Análise Multivariada:
Análise de Componentes Principais (PCA): Reduza a dimensionalidade do dataset para identificar combinações de variáveis que possam ter uma maior correlação com a variável alvo. Isso também pode ajudar a eliminar variáveis irrelevantes.
Clusterização: Utilize técnicas como k-means para agrupar os dados e identificar se certos clusters (grupos de transações) têm maior probabilidade de fraude.

4. Mudança na Definição da Variável Alvo:
Segmentação da Variável Alvo: Em vez de usar Fraude como variável binária, você pode segmentar em diferentes tipos de fraudes (se houver essa distinção nos dados) ou criar subcategorias de fraudes com base em outras características.
Modelo de Anomalia: Utilize a detecção de anomalias como a variável alvo, onde o foco é identificar outliers que diferem significativamente das transações normais, independentemente de serem fraudes confirmadas.

5. Testes A/B e Análises Adicionais:
Testes A/B: Se tiver acesso a dados mais detalhados, realizar testes controlados para entender o impacto de certas variáveis sobre fraudes pode fornecer insights mais profundos.
Coleta de Novos Dados: Considere a possibilidade de coletar ou gerar novos dados que possam estar mais fortemente correlacionados com fraudes, como o comportamento histórico de transações, padrões de geolocalização, etc.
Sugestão para Variáveis Alvo Alternativas:
Se você deseja considerar múltiplas variáveis alvo ou redefinir o conceito de fraude, aqui estão algumas sugestões:

Fraude e Anomalia Temporal: Combine Fraude com a identificação de transações em horários incomuns ou com padrões temporais atípicos.
Fraude e Frequência de Uso: Combine Fraude com uma variável que capture a frequência de uso do cartão em um curto período de tempo.
Fraude por Valor e Localização: Combine Fraude com padrões que envolvem altos valores de transação e mudanças rápidas de localização.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Remover a variável ID_Transacao
df_sem_id = df.drop(columns=['ID_Transacao'])

# Converter todas as variáveis categóricas para numéricas
df_encoded = df_sem_id.copy()
for col in df_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Calcular a matriz de correlação
matriz_correlacao = df_encoded.corr()

# Exibir a matriz de correlação
print(matriz_correlacao)

# Identificar as correlações mais fortes, médias e fracas
correlacoes_fortes = []
correlacoes_medias = []
correlacoes_fracas = []

for i in range(len(matriz_correlacao.columns)):
    for j in range(i + 1, len(matriz_correlacao.columns)):
        coeficiente = matriz_correlacao.iloc[i, j]
        variavel_1 = matriz_correlacao.columns[i]
        variavel_2 = matriz_correlacao.columns[j]
        
        if abs(coeficiente) > 0.7:
            correlacoes_fortes.append((variavel_1, variavel_2, coeficiente))
        elif 0.3 < abs(coeficiente) <= 0.7:
            correlacoes_medias.append((variavel_1, variavel_2, coeficiente))
        else:
            correlacoes_fracas.append((variavel_1, variavel_2, coeficiente))

# Relatório
print("\nCorrelações Fortes (|Coeficiente| > 0.7):")
for item in correlacoes_fortes:
    print(f"{item[0]} e {item[1]}: Coeficiente = {item[2]}")

print("\nCorrelações Médias (0.3 < |Coeficiente| <= 0.7):")
for item in correlacoes_medias:
    print(f"{item[0]} e {item[1]}: Coeficiente = {item[2]}")

print("\nCorrelações Fracas (|Coeficiente| <= 0.3):")
for item in correlacoes_fracas:
    print(f"{item[0]} e {item[1]}: Coeficiente = {item[2]}")
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Selecionar as variáveis numéricas, incluindo as novas variáveis alvo
variaveis_numericas = df[['Valor_Transacao', 'Tentativas_Autenticacao_Falhas', 'Risco_Transacao', 'Fraude_Tipo_Cartao', 'Fraude']]

# Adicionar uma constante para o cálculo do VIF
variaveis_numericas_com_constante = add_constant(variaveis_numericas)

# Calcular o VIF para cada variável
vif = pd.DataFrame()
vif["Variável"] = variaveis_numericas_com_constante.columns
vif["VIF"] = [variance_inflation_factor(variaveis_numericas_com_constante.values, i) for i in range(variaveis_numericas_com_constante.shape[1])]

# Exibir os resultados
print(vif)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Listar as variáveis numéricas que serão analisadas
variaveis_numericas = ['Valor_Transacao', 'Tentativas_Autenticacao_Falhas', 'Risco_Transacao', 'Fraude_Tipo_Cartao', 'Fraude']

# Criar boxplots para cada variável numérica
plt.figure(figsize=(15, 10))
for i, variavel in enumerate(variaveis_numericas, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[variavel])
    plt.title(f'Boxplot de {variavel}')
plt.tight_layout()
plt.show()

# Identificar outliers usando o IQR
outliers = {}
for variavel in variaveis_numericas:
    Q1 = df[variavel].quantile(0.25)
    Q3 = df[variavel].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers[variavel] = df[(df[variavel] < lower_bound) | (df[variavel] > upper_bound)][variavel]

# Exibir os outliers detectados
for variavel, valores_outliers in outliers.items():
    print(f'\nOutliers detectados na variável {variavel}:')
    print(valores_outliers)
```

O fato de 1 ser identificado como outlier pode ser atribuído ao baixo número de fraudes no dataset (desequilíbrio de classes), onde as transações fraudulentas são raras. Portanto, elas aparecem como "outliers" quando comparadas ao grande número de transações não fraudulentas (0).

Conclusão e Próximos Passos:

Desconsiderar Outliers em Variáveis Binárias: Para variáveis como Fraude e Risco_Transacao, que são binárias, os "outliers" identificados representam a ocorrência menos frequente e, portanto, não precisam ser tratados como outliers no sentido tradicional.

Validação de Modelos: Agora que você sabe que não há outliers significativos nas variáveis numéricas principais (exceto Fraude devido ao desequilíbrio de classes), pode seguir com a modelagem, ciente de que a distribuição dos dados não será distorcida por valores anômalos.
