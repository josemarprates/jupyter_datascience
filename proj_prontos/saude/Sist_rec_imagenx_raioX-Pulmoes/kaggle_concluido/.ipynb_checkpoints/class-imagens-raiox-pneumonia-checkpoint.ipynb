```python
# Bibliotecas para manipulação de dados e visualização
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Bibliotecas para machine learning e deep learning
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
```

```python
# Diretórios do dataset no ambiente Kaggle
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# Visualizar a quantidade de imagens em cada classe
def count_images(directory):
    classes = ['NORMAL', 'PNEUMONIA']
    counts = {cls: len(os.listdir(os.path.join(directory, cls))) for cls in classes}
    return counts

# Contagem de imagens no conjunto de treino, validação e teste
train_counts = count_images(train_dir)
val_counts = count_images(val_dir)
test_counts = count_images(test_dir)

print("Imagens no conjunto de treino:", train_counts)
print("Imagens no conjunto de validação:", val_counts)
print("Imagens no conjunto de teste:", test_counts)

# Visualizar as classes em gráficos
sns.barplot(x=list(train_counts.keys()), y=list(train_counts.values()))
plt.title("Distribuição de Imagens no Conjunto de Treino")
plt.ylabel("Número de Imagens")
plt.show()
```

```python
# Criar geradores de dados com Data Augmentation para o treino
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,  # Rotação leve
    width_shift_range=0.2,  # Translação horizontal
    height_shift_range=0.2,  # Translação vertical
    shear_range=0.2,  # Cisalhamento
    zoom_range=0.2,  # Zoom
    horizontal_flip=True,  # Flip horizontal
    fill_mode='nearest'  # Preenchimento de áreas perdidas
)

# Geradores para validação e teste (apenas normalização)
data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Carregar dados de treino
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Carregar dados de validação
val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Carregar dados de teste
test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False  # Não embaralhar para facilitar análise posterior
)
```

```python
# Carregar o modelo pré-treinado EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Congelar camadas base

# Construir o modelo
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu', kernel_regularizer=l2(0.01)),  # Regularização L2
    Dropout(0.4),  # Dropout para evitar overfitting
    Dense(1, activation='sigmoid')  # Saída binária
])

# Compilar o modelo
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Exibir o resumo do modelo
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# EarlyStopping para interromper o treinamento caso não haja melhora
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitorar a perda na validação
    patience=3,  # Parar após 3 épocas sem melhora
    restore_best_weights=True  # Restaurar os melhores pesos do modelo
)

# ReduceLROnPlateau para ajustar dinamicamente a taxa de aprendizado
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',  # Monitorar a perda na validação
    factor=0.5,  # Reduzir a taxa de aprendizado pela metade
    patience=2,  # Esperar 2 épocas antes de reduzir
    min_lr=1e-7  # Limite mínimo para a taxa de aprendizado
)

# Exibir para verificar se foi criado corretamente
print("Callbacks configurados com sucesso.")
```

**INICIO DO TREINAMENTO DO MODELO COM PREPARAÇÃO DOS DADOS NA MESMAS CELULAS**

```python
# Importação de bibliotecas necessárias
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import Recall
from sklearn.metrics import confusion_matrix, classification_report

# 1. Diretórios do Dataset
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# 2. Data Augmentation e Normalização
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,  # Aumentar rotação
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)

data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Gerar os dados
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# 3. Penalização nas Classes (class_weight)
# Obter as proporções de classes no conjunto de treino
train_class_counts = train_data.classes
class_weight = {
    0: len(train_class_counts) / (2 * np.bincount(train_class_counts)[0]),  # Peso para NORMAL
    1: len(train_class_counts) / (2 * np.bincount(train_class_counts)[1])   # Peso para PNEUMONIA
}

# 4. Construção do Modelo com EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True  # Descongelar para Fine-Tuning

model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.3),  # Dropout ajustado
    Dense(1, activation='sigmoid')
])

# Compilar o modelo
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=5e-7),  # Taxa de aprendizado menor
    loss='binary_crossentropy',
    metrics=['accuracy', Recall()]
)

# 5. Callbacks: EarlyStopping com Monitoramento de Recall
early_stopping = EarlyStopping(
    monitor='val_recall',  # Monitorar o recall na validação
    mode='max',  # Maximizar o recall
    patience=5,
    restore_best_weights=True
)

# 6. Treinamento do Modelo com class_weight aplicado
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    steps_per_epoch=train_data.samples // train_data.batch_size,
    validation_steps=val_data.samples // val_data.batch_size,
    class_weight=class_weight,  # Aplicar class_weight para balanceamento
    callbacks=[early_stopping]
)

# 7. Avaliação no Conjunto de Teste
evaluation = model.evaluate(test_data)
loss = evaluation[0]  # Perda
accuracy = evaluation[1]  # Acurácia
recall = evaluation[2]  # Recall

print(f"Loss no teste: {loss}")
print(f"Accuracy no teste: {accuracy}")
print(f"Recall no teste: {recall}")

# 8. Relatório de Classificação
y_true = test_data.classes
y_probs = model.predict(test_data)
y_pred = (y_probs > 0.5).astype(int)

# Matriz de Confusão
cm = confusion_matrix(y_true, y_pred)
print("Matriz de Confusão:")
print(cm)

# Relatório de Classificação
print("Relatório de Classificação:")
print(classification_report(y_true, y_pred, target_names=['NORMAL', 'PNEUMONIA']))
```

A ACIMA TIVEMOS UM BOM RESULTADO MAS AINDA TEMOS ESPAÇOS PARA MELHORA, 
**ABAIXO VAMOS AJUSTAR O THRESSHOLD PARA VER SE TEMOS MELHORA**

```python
# Importação de bibliotecas necessárias
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils import shuffle

# 1. Diretórios do Dataset
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# 2. Data Augmentation e Normalização
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)

data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Gerar os dados
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# 3. Oversampling no conjunto de treino
x_train, y_train = [], []

# Extração de dados do gerador
for _ in range(len(train_data)):
    images, labels = train_data.__next__()
    x_train.extend(images)
    y_train.extend(labels)

x_train, y_train = np.array(x_train), np.array(y_train)

# Aplicar oversampling para balancear as classes
ros = RandomOverSampler()
x_resampled, y_resampled = ros.fit_resample(
    x_train.reshape(x_train.shape[0], -1), y_train)

# Restaurar o formato das imagens
x_resampled = x_resampled.reshape(-1, 224, 224, 3)
y_resampled = y_resampled

# Misturar os dados
x_resampled, y_resampled = shuffle(x_resampled, y_resampled)

# 4. Construção do Modelo
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True

model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.4),  # Dropout ajustado
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-6),  # Taxa de aprendizado reduzida
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 5. Callbacks: EarlyStopping com val_loss
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitorar a perda de validação
    mode='min',  # Minimizar a perda
    patience=8,
    restore_best_weights=True
)

# 6. Treinamento do modelo com oversampled data
history = model.fit(
    x_resampled, y_resampled,
    validation_data=val_data,
    epochs=30,
    batch_size=32,
    callbacks=[early_stopping]
)

# 7. Avaliação no Conjunto de Teste
evaluation = model.evaluate(test_data)
loss, accuracy = evaluation[0], evaluation[1]
print(f"Loss no teste: {loss}")
print(f"Accuracy no teste: {accuracy}")

# 8. Previsões no conjunto de teste com threshold ajustado
y_probs = model.predict(test_data)
y_pred_default = (y_probs > 0.5).astype(int)

# Matriz de Confusão
y_true = test_data.classes
cm_default = confusion_matrix(y_true, y_pred_default)
print("Matriz de Confusão com Threshold Padrão (0.5):")
print(cm_default)

# Relatório de Classificação
print("Relatório de Classificação com Threshold Padrão (0.5):")
print(classification_report(y_true, y_pred_default, target_names=['NORMAL', 'PNEUMONIA']))

# Ajustar o threshold para 0.5
new_threshold = 0.5
y_pred_adjusted = (y_probs > new_threshold).astype(int)

# Matriz de Confusão com Novo Threshold
cm_adjusted = confusion_matrix(y_true, y_pred_adjusted)
print("Matriz de Confusão com Novo Threshold (0.6):")
print(cm_adjusted)

# Relatório de Classificação com Novo Threshold
print("Relatório de Classificação com Novo Threshold (0.6):")
print(classification_report(y_true, y_pred_adjusted, target_names=['NORMAL', 'PNEUMONIA']))

# 9. Visualização do Treinamento
import matplotlib.pyplot as plt

# Acurácia
plt.plot(history.history['accuracy'], label='Acurácia de Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia durante o Treinamento')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
plt.show()

# Perda
plt.plot(history.history['loss'], label='Perda de Treino')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda durante o Treinamento')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.show()
```

Relatório Final do Projeto de Detecção de Pneumonia em Raios-X
Resumo do Projeto
O objetivo deste projeto foi desenvolver e treinar um modelo de aprendizado profundo utilizando a rede pré-treinada EfficientNetB0 para classificar imagens de raios-X em duas categorias:

NORMAL - Pulmões sem sinais de pneumonia.
PNEUMONIA - Pulmões com sinais de pneumonia.
O modelo foi avaliado com métricas relevantes ao contexto médico, priorizando a identificação correta de casos de pneumonia para reduzir falsos negativos, enquanto se buscava manter um equilíbrio geral no desempenho.

Pipeline do Modelo
Pré-processamento dos Dados:

As imagens foram redimensionadas para 224x224 pixels.
Data Augmentation foi aplicado no conjunto de treino, incluindo:
Rotações, translações, zooms e flips horizontais para aumentar a variabilidade dos dados.
Normalização foi realizada dividindo os valores dos pixels por 255, escalando-os para o intervalo [0, 1].
Balanceamento de Classes:

Utilizou-se Random Oversampling para balancear o dataset. A classe minoritária (NORMAL) foi duplicada no espaço de características para igualar o número de amostras com a classe majoritária (PNEUMONIA).
Modelo Utilizado:

A rede EfficientNetB0, pré-treinada no ImageNet, foi utilizada como extrator de características, com sua parte convolucional descongelada para fine-tuning.
Camadas densas adicionais foram adicionadas:
Camada densa de 512 unidades com ativação ReLU.
Dropout de 40% para evitar overfitting.
Camada final com ativação sigmoidal para previsão binária.
Treinamento:

O modelo foi treinado por até 30 épocas com a técnica de early stopping baseada no monitoramento da perda de validação (val_loss), interrompendo após 8 épocas sem melhoria.
Uma taxa de aprendizado reduzida de 1e-6 foi utilizada para garantir estabilidade durante o fine-tuning.
Avaliação:

O modelo foi avaliado no conjunto de teste com threshold padrão (0.5) e um threshold ajustado.
Métricas como precision, recall, F1-Score e matriz de confusão foram analisadas.
