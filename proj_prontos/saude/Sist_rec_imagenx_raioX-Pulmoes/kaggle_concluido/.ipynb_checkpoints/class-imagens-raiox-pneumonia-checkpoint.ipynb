```python
# Bibliotecas para manipulao de dados e visualizao
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Bibliotecas para machine learning e deep learning
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
```

```python
# Diretrios do dataset no ambiente Kaggle
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# Visualizar a quantidade de imagens em cada classe
def count_images(directory):
    classes = ['NORMAL', 'PNEUMONIA']
    counts = {cls: len(os.listdir(os.path.join(directory, cls))) for cls in classes}
    return counts

# Contagem de imagens no conjunto de treino, validao e teste
train_counts = count_images(train_dir)
val_counts = count_images(val_dir)
test_counts = count_images(test_dir)

print("Imagens no conjunto de treino:", train_counts)
print("Imagens no conjunto de validao:", val_counts)
print("Imagens no conjunto de teste:", test_counts)

# Visualizar as classes em grficos
sns.barplot(x=list(train_counts.keys()), y=list(train_counts.values()))
plt.title("Distribuio de Imagens no Conjunto de Treino")
plt.ylabel("Nmero de Imagens")
plt.show()
```

```python
# Criar geradores de dados com Data Augmentation para o treino
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,  # Rotao leve
    width_shift_range=0.2,  # Translao horizontal
    height_shift_range=0.2,  # Translao vertical
    shear_range=0.2,  # Cisalhamento
    zoom_range=0.2,  # Zoom
    horizontal_flip=True,  # Flip horizontal
    fill_mode='nearest'  # Preenchimento de reas perdidas
)

# Geradores para validao e teste (apenas normalizao)
data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Carregar dados de treino
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Carregar dados de validao
val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Carregar dados de teste
test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False  # No embaralhar para facilitar anlise posterior
)
```

```python
# Carregar o modelo pr-treinado EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Congelar camadas base

# Construir o modelo
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu', kernel_regularizer=l2(0.01)),  # Regularizao L2
    Dropout(0.4),  # Dropout para evitar overfitting
    Dense(1, activation='sigmoid')  # Sada binria
])

# Compilar o modelo
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Exibir o resumo do modelo
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# EarlyStopping para interromper o treinamento caso no haja melhora
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitorar a perda na validao
    patience=3,  # Parar aps 3 pocas sem melhora
    restore_best_weights=True  # Restaurar os melhores pesos do modelo
)

# ReduceLROnPlateau para ajustar dinamicamente a taxa de aprendizado
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',  # Monitorar a perda na validao
    factor=0.5,  # Reduzir a taxa de aprendizado pela metade
    patience=2,  # Esperar 2 pocas antes de reduzir
    min_lr=1e-7  # Limite mnimo para a taxa de aprendizado
)

# Exibir para verificar se foi criado corretamente
print("Callbacks configurados com sucesso.")
```

**INICIO DO TREINAMENTO DO MODELO COM PREPARAO DOS DADOS NA MESMAS CELULAS**

```python
# Importao de bibliotecas necessrias
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import Recall
from sklearn.metrics import confusion_matrix, classification_report

# 1. Diretrios do Dataset
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# 2. Data Augmentation e Normalizao
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,  # Aumentar rotao
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)

data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Gerar os dados
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# 3. Penalizao nas Classes (class_weight)
# Obter as propores de classes no conjunto de treino
train_class_counts = train_data.classes
class_weight = {
    0: len(train_class_counts) / (2 * np.bincount(train_class_counts)[0]),  # Peso para NORMAL
    1: len(train_class_counts) / (2 * np.bincount(train_class_counts)[1])   # Peso para PNEUMONIA
}

# 4. Construo do Modelo com EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True  # Descongelar para Fine-Tuning

model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.3),  # Dropout ajustado
    Dense(1, activation='sigmoid')
])

# Compilar o modelo
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=5e-7),  # Taxa de aprendizado menor
    loss='binary_crossentropy',
    metrics=['accuracy', Recall()]
)

# 5. Callbacks: EarlyStopping com Monitoramento de Recall
early_stopping = EarlyStopping(
    monitor='val_recall',  # Monitorar o recall na validao
    mode='max',  # Maximizar o recall
    patience=5,
    restore_best_weights=True
)

# 6. Treinamento do Modelo com class_weight aplicado
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    steps_per_epoch=train_data.samples // train_data.batch_size,
    validation_steps=val_data.samples // val_data.batch_size,
    class_weight=class_weight,  # Aplicar class_weight para balanceamento
    callbacks=[early_stopping]
)

# 7. Avaliao no Conjunto de Teste
evaluation = model.evaluate(test_data)
loss = evaluation[0]  # Perda
accuracy = evaluation[1]  # Acurcia
recall = evaluation[2]  # Recall

print(f"Loss no teste: {loss}")
print(f"Accuracy no teste: {accuracy}")
print(f"Recall no teste: {recall}")

# 8. Relatrio de Classificao
y_true = test_data.classes
y_probs = model.predict(test_data)
y_pred = (y_probs > 0.5).astype(int)

# Matriz de Confuso
cm = confusion_matrix(y_true, y_pred)
print("Matriz de Confuso:")
print(cm)

# Relatrio de Classificao
print("Relatrio de Classificao:")
print(classification_report(y_true, y_pred, target_names=['NORMAL', 'PNEUMONIA']))
```

A ACIMA TIVEMOS UM BOM RESULTADO MAS AINDA TEMOS ESPAOS PARA MELHORA, 
**ABAIXO VAMOS AJUSTAR O THRESSHOLD PARA VER SE TEMOS MELHORA**

```python
# Importao de bibliotecas necessrias
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils import shuffle

# 1. Diretrios do Dataset
train_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/train'
val_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/val'
test_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray/test'

# 2. Data Augmentation e Normalizao
data_gen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)

data_gen_val_test = ImageDataGenerator(rescale=1./255)

# Gerar os dados
train_data = data_gen_train.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = data_gen_val_test.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

test_data = data_gen_val_test.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# 3. Oversampling no conjunto de treino
x_train, y_train = [], []

# Extrao de dados do gerador
for _ in range(len(train_data)):
    images, labels = train_data.__next__()
    x_train.extend(images)
    y_train.extend(labels)

x_train, y_train = np.array(x_train), np.array(y_train)

# Aplicar oversampling para balancear as classes
ros = RandomOverSampler()
x_resampled, y_resampled = ros.fit_resample(
    x_train.reshape(x_train.shape[0], -1), y_train)

# Restaurar o formato das imagens
x_resampled = x_resampled.reshape(-1, 224, 224, 3)
y_resampled = y_resampled

# Misturar os dados
x_resampled, y_resampled = shuffle(x_resampled, y_resampled)

# 4. Construo do Modelo
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True

model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.4),  # Dropout ajustado
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-6),  # Taxa de aprendizado reduzida
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 5. Callbacks: EarlyStopping com val_loss
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitorar a perda de validao
    mode='min',  # Minimizar a perda
    patience=8,
    restore_best_weights=True
)

# 6. Treinamento do modelo com oversampled data
history = model.fit(
    x_resampled, y_resampled,
    validation_data=val_data,
    epochs=30,
    batch_size=32,
    callbacks=[early_stopping]
)

# 7. Avaliao no Conjunto de Teste
evaluation = model.evaluate(test_data)
loss, accuracy = evaluation[0], evaluation[1]
print(f"Loss no teste: {loss}")
print(f"Accuracy no teste: {accuracy}")

# 8. Previses no conjunto de teste com threshold ajustado
y_probs = model.predict(test_data)
y_pred_default = (y_probs > 0.5).astype(int)

# Matriz de Confuso
y_true = test_data.classes
cm_default = confusion_matrix(y_true, y_pred_default)
print("Matriz de Confuso com Threshold Padro (0.5):")
print(cm_default)

# Relatrio de Classificao
print("Relatrio de Classificao com Threshold Padro (0.5):")
print(classification_report(y_true, y_pred_default, target_names=['NORMAL', 'PNEUMONIA']))

# Ajustar o threshold para 0.5
new_threshold = 0.5
y_pred_adjusted = (y_probs > new_threshold).astype(int)

# Matriz de Confuso com Novo Threshold
cm_adjusted = confusion_matrix(y_true, y_pred_adjusted)
print("Matriz de Confuso com Novo Threshold (0.6):")
print(cm_adjusted)

# Relatrio de Classificao com Novo Threshold
print("Relatrio de Classificao com Novo Threshold (0.6):")
print(classification_report(y_true, y_pred_adjusted, target_names=['NORMAL', 'PNEUMONIA']))

# 9. Visualizao do Treinamento
import matplotlib.pyplot as plt

# Acurcia
plt.plot(history.history['accuracy'], label='Acurcia de Treino')
plt.plot(history.history['val_accuracy'], label='Acurcia de Validao')
plt.title('Acurcia durante o Treinamento')
plt.xlabel('pocas')
plt.ylabel('Acurcia')
plt.legend()
plt.show()

# Perda
plt.plot(history.history['loss'], label='Perda de Treino')
plt.plot(history.history['val_loss'], label='Perda de Validao')
plt.title('Perda durante o Treinamento')
plt.xlabel('pocas')
plt.ylabel('Perda')
plt.legend()
plt.show()
```

Relatrio Final do Projeto de Deteco de Pneumonia em Raios-X
Resumo do Projeto
O objetivo deste projeto foi desenvolver e treinar um modelo de aprendizado profundo utilizando a rede pr-treinada EfficientNetB0 para classificar imagens de raios-X em duas categorias:

NORMAL - Pulmes sem sinais de pneumonia.
PNEUMONIA - Pulmes com sinais de pneumonia.
O modelo foi avaliado com mtricas relevantes ao contexto mdico, priorizando a identificao correta de casos de pneumonia para reduzir falsos negativos, enquanto se buscava manter um equilbrio geral no desempenho.

Pipeline do Modelo
Pr-processamento dos Dados:

As imagens foram redimensionadas para 224x224 pixels.
Data Augmentation foi aplicado no conjunto de treino, incluindo:
Rotaes, translaes, zooms e flips horizontais para aumentar a variabilidade dos dados.
Normalizao foi realizada dividindo os valores dos pixels por 255, escalando-os para o intervalo [0, 1].
Balanceamento de Classes:

Utilizou-se Random Oversampling para balancear o dataset. A classe minoritria (NORMAL) foi duplicada no espao de caractersticas para igualar o nmero de amostras com a classe majoritria (PNEUMONIA).
Modelo Utilizado:

A rede EfficientNetB0, pr-treinada no ImageNet, foi utilizada como extrator de caractersticas, com sua parte convolucional descongelada para fine-tuning.
Camadas densas adicionais foram adicionadas:
Camada densa de 512 unidades com ativao ReLU.
Dropout de 40% para evitar overfitting.
Camada final com ativao sigmoidal para previso binria.
Treinamento:

O modelo foi treinado por at 30 pocas com a tcnica de early stopping baseada no monitoramento da perda de validao (val_loss), interrompendo aps 8 pocas sem melhoria.
Uma taxa de aprendizado reduzida de 1e-6 foi utilizada para garantir estabilidade durante o fine-tuning.
Avaliao:

O modelo foi avaliado no conjunto de teste com threshold padro (0.5) e um threshold ajustado.
Mtricas como precision, recall, F1-Score e matriz de confuso foram analisadas.
