```python
# Função para verificar a presença de outliers usando o método IQR
def detect_outliers(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers_present = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        if outliers_present > 0:
            outliers[col] = outliers_present
    return outliers

# Verificar a presença de outliers nas variáveis numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns
outliers = detect_outliers(df, numeric_columns)

# Exibir as variáveis que possuem outliers
outliers
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Definir as variáveis preditoras categóricas e numéricas
categorical_columns = ['payment_method', 'contract', 'online_security', 'tech_support', 'partner', 'dependents']
numerical_columns = ['senior_citizen', 'total_charges', 'monthly_charges', 'tenure']

# Separar as variáveis preditoras e a variável alvo
X = df[categorical_columns + numerical_columns]
y = df['Churn']

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Aplicar a codificação nas variáveis categóricas e padronizar as variáveis numéricas
column_transformer = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_columns),
    ('num', StandardScaler(), numerical_columns)
])

X_train_transformed = column_transformer.fit_transform(X_train)
X_test_transformed = column_transformer.transform(X_test)

# Treinar o modelo XGBoost
model = XGBClassifier(random_state=42, n_estimators=50, eval_metric='logloss')
model.fit(X_train_transformed, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test_transformed)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Exibir os resultados
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from xgboost import XGBClassifier

# Definir as variáveis preditoras categóricas e numéricas
categorical_columns = ['payment_method', 'contract', 'online_security', 'tech_support', 'partner', 'dependents']
numerical_columns = ['senior_citizen', 'total_charges', 'monthly_charges', 'tenure']

# Separar as variáveis preditoras e a variável alvo
X = df[categorical_columns + numerical_columns]
y = df['Churn']

# Aplicar a codificação nas variáveis categóricas e padronizar as variáveis numéricas
column_transformer = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_columns),
    ('num', StandardScaler(), numerical_columns)
])

X_transformed = column_transformer.fit_transform(X)

# Aplicar SMOTE para balancear as classes após a codificação
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_transformed, y)

# Dividir os dados balanceados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Ajuste de hiperparâmetros usando GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3]
}

xgb = XGBClassifier(random_state=42, eval_metric='logloss')
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Melhor modelo após o ajuste de hiperparâmetros
best_model = grid_search.best_estimator_

# Fazer previsões no conjunto de teste
y_pred = best_model.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Exibir os resultados
print(f'Best Parameters: {grid_search.best_params_}')
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from xgboost import XGBClassifier

# Definir as novas variáveis preditoras categóricas e numéricas
categorical_columns = ['payment_method', 'contract', 'online_security', 'tech_support', 'partner', 'dependents']
numerical_columns = ['senior_citizen', 'total_charges', 'monthly_charges', 'tenure']

# Separar as variáveis preditoras e a variável alvo
X = df[categorical_columns + numerical_columns]
y = df['Churn']

# Aplicar a codificação nas variáveis categóricas e padronizar as variáveis numéricas
column_transformer = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_columns),
    ('num', StandardScaler(), numerical_columns)
])

X_transformed = column_transformer.fit_transform(X)

# Aplicar SMOTE para balancear as classes após a codificação
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_transformed, y)

# Dividir os dados balanceados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Manter os hiperparâmetros encontrados no código anterior
best_params = {
    'learning_rate': 0.3,
    'max_depth': 3,
    'n_estimators': 150
}

# Treinar o modelo XGBoost com os melhores hiperparâmetros
model = XGBClassifier(**best_params, random_state=42, eval_metric='logloss')
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Exibir os resultados
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier

# Definir as variáveis preditoras categóricas e numéricas, incluindo a nova variável
categorical_columns = ['payment_method', 'contract', 'online_security', 'tech_support', 'partner', 'dependents']
numerical_columns = ['senior_citizen', 'total_charges', 'monthly_charges', 'tenure', 'growth_rate_of_cost']

# Separar as variáveis preditoras e a variável alvo
X = df[categorical_columns + numerical_columns]
y = df['Churn']

# Aplicar a codificação nas variáveis categóricas e padronizar as variáveis numéricas
column_transformer = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_columns),
    ('num', StandardScaler(), numerical_columns)
])

X_transformed = column_transformer.fit_transform(X)

# Aplicar SMOTE para balancear as classes após a codificação
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_transformed, y)

# Dividir os dados balanceados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Manter os mesmos hiperparâmetros usados anteriormente
best_params = {
    'learning_rate': 0.3,
    'max_depth': 3,
    'n_estimators': 150,
    'n_jobs': 1  # Forçar o XGBoost a usar uma única thread
}

# Treinar o modelo XGBoost com a nova variável
model = XGBClassifier(**best_params, random_state=42, eval_metric='logloss')
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Exibir os resultados
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
